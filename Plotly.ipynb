# Plotly

Scatter Plot

```python
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go

#The code below allows plotly to render plots inline using the iplot function
pyo.init_notebook_mode(connected=True)

np.random.seed(42)

random_x = np.random.randint(1,101,100)
random_y= np.random.randint(1,101,100)
```

```python
data= [go.Scatter(x=random_x,
                  y=random_y,
                  mode='markers',
                 marker = dict(
                     size=12,
                     color='rgb(51,204,153)',
                     symbol='pentagon',
                     line ={'width':2}
                 ))]
layout = go.Layout(title ='Hello First Plot',
                   xaxis = {'title': 'My X Axis'},
                   yaxis = dict(title ='My Y Axis'),
                   hovermode ='closest')

fig = go.Figure(data=data, layout = layout)
pyo.iplot(fig, filename ='scatter.html')
```

Line Chart

```python
x_values = np.linspace(0,1,100)
y_values = np.random.randn(100)

trace0 = go.Scatter(x=x_values, y=y_values+5, 
                  mode= 'markers', name ='markers')

trace1 = go.Scatter(x=x_values, y=y_values,
                   mode ='lines', name ='mylines')

trace2 = go.Scatter(x=x_values, y=y_values-5,
                   mode ='lines+markers', name ='myfav')

data = [trace0, trace1, trace2]

layout =go.Layout(title ='Line Charts')

fig = go.Figure(data =data, layout=layout)

pyo.iplot(fig, filename='Line Chart.html')
```

Bar Charts

```python
df = pd.read_csv('../Data/2018WinterOlympics.csv')
```

```python
df
```

```python
data = [go.Bar(x= df['NOC'], y=df['Total'])]
layout=go.Layout(title='Medals',
                xaxis=dict(title='Country'),
                yaxis=dict(title= 'Number of medals'))
fig =go.Figure(data =data, layout =layout)

pyo.iplot(fig)
```

```python
#plot a nested bar chart, we need to define three traces corresponding to the different medal categories
trace1= go.Bar(x= df['NOC'], y=df['Gold'], name='Gold', marker=dict(color='#FFD700'))
trace2= go.Bar(x= df['NOC'], y=df['Silver'], name='Silver', marker=dict(color='#9EA0A1'))
trace3= go.Bar(x= df['NOC'], y=df['Bronze'], name='Bronze', marker=dict(color='#CD7F32'))

data = [trace1, trace2, trace3]

layout=go.Layout(title='Medals',
                xaxis=dict(title='Country'),
                yaxis=dict(title= 'Number of medals'))
fig =go.Figure(data =data, layout =layout)

pyo.iplot(fig)
```

```python
#make a nested barplot
trace1= go.Bar(x= df['NOC'], y=df['Gold'], name='Gold', marker=dict(color='#FFD700'))
trace2= go.Bar(x= df['NOC'], y=df['Silver'], name='Silver', marker=dict(color='#9EA0A1'))
trace3= go.Bar(x= df['NOC'], y=df['Bronze'], name='Bronze', marker=dict(color='#CD7F32'))

data = [trace1, trace2, trace3]
# to change the plot, we add an extra arguement called 'barmode' into the layout function
layout=go.Layout(title='Medals',
                xaxis=dict(title='Country'),
                yaxis=dict(title= 'Number of medals'),
                barmode='stack')
fig =go.Figure(data =data, layout =layout)

pyo.iplot(fig)
```

Bubble Charts

```python
df = pd.read_csv('../data/mpg.csv')
```

```python
df
```

```python
data = [go.Scatter(x=df['horsepower'], 
                   y=df['mpg'],
                  text =df['name'],
                  mode ='markers',
                  marker =dict(size=df['weight']/200, 
                               color=df['cylinders'],
                              showscale=True))]

layout = go.Layout(title='Buble Chart',
                  xaxis=dict(title='Horsepower'),
                  yaxis=dict(title='MPG'))
fig = go.Figure(data=data, layout =layout)
pyo.iplot(fig)
```
