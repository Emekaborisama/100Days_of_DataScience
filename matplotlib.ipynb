<h1> <b>Matplotlib </b> </h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#linspace returns evenly spaced numbers over a specified interval. Here the start is 0, stop is 5 and the number of steps is 11
x = np.linspace(0,5,11)
y = x **2
```

```python
print(x)
print(y)
```

```python
#Functional method of plotting
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

# multiplots on the same canvas

```python
#for subplots, the first arguements is the number of rows, second arguement is the number of columns and the third is the plot number
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# the object oriented way for plotting

```python
# here one can create figure objects and then call methods of it
fig =plt.figure()

# the above code creates a sort of imaginary canvas
# we pass in more information (axes) to define the figure. to define the axes, we pass in a list of four numbers, Left, bottom,
# width and height

axes =fig.add_axes([0.1,0.1,0.8,0.8])
# after defining the axes, we then plot on the axes

axes.plot(x,y)
# we then customise the axes

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig =plt.figure()

axes1 =fig.add_axes([0.1,0.1,0.8,0.8])
axes2 =fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Big Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig,axes =plt.subplots(1,2)
#the axes is a list of axes, one can iterate over it or call each one

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout() #this takes care of overlaps
```

# Figuresize and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Customisation

```python
fig = plt.figure()

ax =fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green')
```

```python
#Line width and Line style
#pass in a value into linewidth or lw
#alpha controls transparency
#linestyle oe ls to define the style of a line
# marker to specify where the points occur

fig = plt.figure()

ax =fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=2, alpha= 0.5, linestyle='--', marker='o', markersize=20 )
```
