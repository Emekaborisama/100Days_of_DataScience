```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].groupby('country').mean()
```

```python
#you can pass in multiple aggregates at once using the agg method
drinks.groupby('continent').beer_servings.agg(['count','mean','std','min','max'])
```

```python
drinks.groupby('continent').beer_servings.mean().plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Average Beer Servings')
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Average')
```

```python
drinks.groupby('continent')[['beer_servings', 'spirit_servings']].mean().plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Average')
```
