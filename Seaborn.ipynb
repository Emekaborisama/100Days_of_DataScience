# Distribution Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
# seaborn  has some datasets preloaded, lets load tips dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.distplot(tips['total_bill'], hist= False)
```

```python
# The joint plot allows one to combine to dist plots
sns.jointplot(x='total_bill', y='tip', data= tips, kind='reg')
```

```python
#pairplot plots pairwise relationship across the entire dataframe (numerical variables)
sns.pairplot(tips, hue= 'sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator= np.std)
```

```python
sns.countplot(x='sex', data =tips)
```

```python
sns.boxplot(x= 'day', y='total_bill', data= tips)
```

```python
sns.violinplot(x= 'day', y='total_bill', data= tips)
```

```python
sns.violinplot(x= 'day', y='total_bill', hue='sex', data= tips, split=True)
```

```python
sns.stripplot(x= 'day', y='total_bill', data =tips, jitter = True)
```

```python
#Swarm plot = violin plot + stripplot. Not ideal with large datasets
sns.swarmplot(x='day', y='total_bill',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data= tips)
sns.swarmplot(x='day', y='total_bill',data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data= tips, kind= 'point')
```

# Matrix Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

to plot a mtrix plot, one needs to first create a matrix.

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot =True)
```

```python
import matplotlib.pyplot as plt
```

```python
fp = flights.pivot_table(index= 'month', columns='year', values='passengers')
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,6))
sns.heatmap(fp, ax=ax, linecolor ='white', linewidths=3)
```

Cluster Map

The cluster map tries to cluster information based on similarity

```python
sns.clustermap(fp, cmap='coolwarm')
```

<h2>GRIDS</h2>

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
#PairGrid gives you control over the pairplot
sns.PairGrid(iris)
```

```python
g= sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

<h3>Facet Grid</h3>

```python
h = sns.FacetGrid(data= tips, col = 'time', row= 'smoker')
```

```python
h = sns.FacetGrid(data= tips, col = 'time', row= 'smoker')
h.map(sns.distplot, 'total_bill')
```

```python
h = sns.FacetGrid(data= tips, col = 'time', row= 'smoker')
h.map(plt.scatter, 'total_bill', 'tip')
```
